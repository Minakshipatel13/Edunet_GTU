##**Numpy**

NumPy (Numerical Python) is a Python library used for fast numerical computations and working with multidimensional arrays.

```python
import numpy as np
```

##Create a Multidimensional Array

```python
arr = np.array([[10, 20, 30],
                [40, 50, 60],
                [70, 80, 90]])

print(arr)
```

##Access Values Using Index Numbers

```python
print("Element at (0,1):", arr[0,1])

print("Element at (2,2):", arr[2,2])

print("Element at (1,0):", arr[1,0])
```

##Mathematical Operations

```python
print("Addition:")
print(arr + 10)

print("\nSubtraction:")
print(arr - 5)

print("\nMultiplication:")
print(arr * 2)

print("\nDivision:")
print(arr / 10)
```
